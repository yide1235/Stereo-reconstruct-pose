```python
import pickle
import numpy as np
```

```python
with open("keypoints.pkl","rb") as f:
    points=pickle.load(f)
with open("boxes.pkl","rb") as f:
    boxes=pickle.load(f)
```

```python
def get_person_vecs_from_points(points):
    vec=[]
    vec_inds=[[5,6],[5,7],[5,11],[6,8],[6,12],[7,9],[8,10],[11,12],[11,13],[12,14],[13,15],[14,16]]
    for pair in vec_inds:
        vec.append(np.linalg.norm(points[pair[0]][0:2]-points[pair[1]][0:2]))
    face_score=min(points[0:5][3])**2*np.linalg.norm(points[2][0:2]-points[3][0:2])
    return vec, face_score
```

```python
image_width=1920
image_height=1080
vecs=[]
face_scores=[]
for i in range(len(points)):
    vec=[]
    box=[]
    bad=False
    for j in range(len(points[i])):
        for k in range(len(points[i][j])):
            if points[i][j][k][0]<0 or points[i][j][k][1]<0 or points[i][j][k][0]>image_width or points[i][j][k][1]>image_height:
                bad=True
        v,f=get_person_vecs_from_points(points[i][j])
        vec.append(v)
    if bad !=True:
        face_scores.append(f)
        vecs.append(vec)
    
dat=[vecs,boxes,face_scores]
```

```python
with open("frames_data_school.pkl","wb") as f:
    pickle.dump(dat,f)
```
